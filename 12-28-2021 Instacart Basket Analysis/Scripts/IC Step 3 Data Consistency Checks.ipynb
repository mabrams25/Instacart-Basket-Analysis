# 4. Check for Null Values each table

```python
# Returns the number of empty values for each column of df_orders, only one column has null values: days_since prior_order

df_orders.isnull().sum()
```

```python
# Seems that whenever someone places their order for the first time, the number of days since prior order is marked not applicable
# There are 206209 observations where people place their first order

df_orders[df_orders['days_since_prior_order'].isnull() == True]
```

```python
# Creating a new column for for orders.csv. 
# We don't want to impute values because it was each customer's first time making an order
# We also don't want to delete the Nan values since we want to use first orders as part of our analysis

df_ords_clean = df_orders
df_ords_clean['first_order'] = df_orders['days_since_prior_order'].isnull() == True
df_ords_clean
```

```python
# Returns the number of empty values for each column, only one column has null values: product_id
df_products.isnull().sum()
```

```python
# Creating a new column for for products.csv. 
# We don't want to impute values because the values there are strings
# We also don't want to delete the Nan values since we want to know who bought the items that happen to be unnamed

df_prods_clean = df_products
df_prods_clean['unknown_product_name'] = df_products['product_name'].isnull() == True
df_prods_clean
```

```python
df_departs.isnull().sum()
```

```python
# Returned columns with missing values, only column is First Name

df_cust.isnull().sum()
```

```python
# Creating a new column for for customers.csv. 
# We don't want to impute values because the values there are strings
# We also don't want to delete the Nan values since we want to be inclusive of all customers making purchases

df_cust_clean = df_cust
df_cust_clean['Missing_First_Name'] = df_cust['First Name'].isnull() == True
df_cust_clean
```

# 5. Check for Data consistency

```python
# Stat description of products dataframe
# Maximum value for price of an item is $99999, which is unusual

df_prods_clean.describe()
```

```python
# Retrieving top 10 highest price amounts
# Two price values $99999 and $14900 are unusual 

df_prods_clean['prices'].nlargest(10)
```

```python
# Returns the name of the products that cost unusually expensive

df_prods_clean[df_prods_clean['prices'] >= 14900]
```

```python
# Returns the name of the alternative products similar to those that are marked extremely high

df_prods_clean.loc[df_prods_clean['product_name'].isin(['Low Fat Cottage Cheese', 'Reduced Fat Milk'])]
```

```python
# Replaces the values of the extremely high prices with the alternative prices under the prices column)

df_prods_clean['prices'].replace({99999: 9.2, 14900: 6.7}, inplace = True)
df_prods_clean.describe()
```
