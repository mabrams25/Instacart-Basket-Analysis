# 3. Aggregating stats regarding number of orders grouped by department

```python
# Creating a dataframe subset with the first 1000000 rows
df_subset = df_merged[:1000000]
df_subset
```

```python
# Finding the mean of the number of orders grouped by department for the first 1000000 rows
# The household department has the highest average number of orders 

df_subset.groupby('department_id')['order_number'].mean().round().astype('Int64')
```

```python
# Finding the mean of the number of orders grouped by department for the entire data frame, rounding to the nearest order
# Here this time we're inclusive of all types of departments. 
# The missing department has the highest average number of orders with approximately 23 orders 

df_merged.groupby('department_id')['order_number'].mean().round().astype('Int64')
```

```python
df_merged.groupby('department_id').agg({'order_number':['mean']}).round().astype('Int64')
```

# 4. Aggregating maximum number of orders grouped by customers. Deriving new column 'loyalty_flag' from max_orders column

```python
# Transforming new column: max_orders based on aggregating the maximum number of orders from each customer

df_merged['max_orders'] = df_merged.groupby(['user_id'])['order_number'].transform(np.max)
df_merged
```

```python
# Deriving new column: loyalty flag based on the max_orders column 
# If max_orders > 40, loyal customer
# If max_orders > 10 but <= 40, regular customer
# If max_orders <= 10 , new customer

df_merged.loc[df_merged['max_orders'] > 40, 'loyalty_flag'] = 'Loyal Customer'
df_merged.loc[(df_merged['max_orders'] <= 40) & (df_merged['max_orders'] > 10), 'loyalty_flag'] = 'Regular Customer'
df_merged.loc[df_merged['max_orders'] <= 10, 'loyalty_flag'] = 'New Customer'
df_merged
```

```python
# The average prices of products purchased by loyal customers barely is barely less than those purchased by regular or new customers
df_merged.groupby('loyalty_flag').agg({'prices':['mean','min','max']}).round(2)
```

# 5. Aggregating average price of products grouped by customers. Deriving new column 'type_of_spender' from average_price_products_purchased column

```python
# Transforming new column: average_price_products_purchased based on aggregating the average price of products spent by customer

df_merged['average_price_products_purchased'] = df_merged.groupby(['user_id'])['prices'].transform(np.mean).round(2)
df_merged
```

```python
# Deriving new column: type_of_spender based on the average_price_products_purchased column 
# If average_price_producsts_purchased < 10, low spender
# If average_price_producsts_purchased >= 10, high spender

df_merged.loc[df_merged['average_price_products_purchased'] < 10, 'type_of_spender'] = 'Low spender'
df_merged.loc[df_merged['average_price_products_purchased'] >= 10, 'type_of_spender'] = 'High spender'
df_merged
```

# 6. Aggregating median_number_of_days grouped by customers. Deriving new column 'type_of_customer' based on average price of products customers spent

```python
# Transforming new column: median_of_days_since_prior_order based on aggregating the median of the number of days customers returned since prior order 

df_merged['median_num_of_days_since_prior_order'] = df_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median).round().astype('Int64')
df_merged
```

```python
# Deriving new column: type_of_customer based on the median_num_of_days_since_prior_order column
# If median_num_of_days_since_prior_order column > 40, Frequent customer
# If median_num_of_days_since_prior_order column > 10 but <= 20, Semi-frequent customer
# If median_num_of_days_since_prior_order column <= 10, Non-frequent customer

df_merged.loc[df_merged['median_num_of_days_since_prior_order'] > 20, 'type_of_customer'] = 'Non-frequent customer'
df_merged.loc[(df_merged['median_num_of_days_since_prior_order'] > 10) & (df_merged['median_num_of_days_since_prior_order'] <= 20), 'type_of_customer'] = 'Semi-frequent customer'  
df_merged.loc[df_merged['median_num_of_days_since_prior_order'] <= 10, 'type_of_customer'] = 'Frequent customer'
df_merged
```
